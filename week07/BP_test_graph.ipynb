```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import matplotlib as mpl

# ✅ 한글 폰트 설정
plt.rcParams['font.family'] = 'Malgun Gothic'  # 윈도우: 맑은 고딕
mpl.rcParams['axes.unicode_minus'] = False

# 데이터 불러오기
df = pd.read_csv("C:/Users/jinwo/Downloads/BP_data.csv")  # 파일 경로는 필요에 따라 수정
```

```python
# 성별 비율 (0: 여성, 1: 남성)
sex_counts = df['Sex'].value_counts()

# 고혈압 여부 비율 (0: 정상, 1: 고혈압)
bp_counts = df['Blood_Pressure_Abnormality'].value_counts()

# 시각화
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# 성별 그래프
axs[0].bar(sex_counts.index.astype(str), sex_counts.values, color='skyblue')
axs[0].set_title("성별 분포")
axs[0].set_xlabel("성별 (0: 여성, 1: 남성)")
axs[0].set_ylabel("명 수")

# 고혈압 여부 그래프
axs[1].bar(bp_counts.index.astype(str), bp_counts.values, color='salmon')
axs[1].set_title("고혈압 여부 분포")
axs[1].set_xlabel("0 = 정상, 1 = 고혈압")
axs[1].set_ylabel("명 수")

plt.tight_layout()
plt.show()
```

```python
# 필요한 열만, NaN 제거
df = df[['Sex', 'Physical_activity']].dropna()

# 성별 그룹
female_activity = df[df['Sex'] == 0]['Physical_activity']
male_activity = df[df['Sex'] == 1]['Physical_activity']

# 박스플롯 생성
plt.figure(figsize=(8, 6))
box = plt.boxplot(
    [female_activity, male_activity],
    patch_artist=True,  # 색상 채우기 활성화
    tick_labels=['여성', '남성']
)

# 색상 설정
colors = ['lightsalmon', 'skyblue']
for patch, color in zip(box['boxes'], colors):
    patch.set_facecolor(color)

plt.title("성별에 따른 Physical_activity 분포")
plt.ylabel("활동량 지수")
plt.grid(True)
plt.show()
```
