```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib as mpl

# 한글 폰트 설정
plt.rcParams['font.family'] = 'Malgun Gothic'
mpl.rcParams['axes.unicode_minus'] = False
```

```python
# 데이터셋 인스턴스 생성

df = pd.read_csv("C:/Users/jinwo/Downloads/BP_data.csv")  # CSV 파일 경로
```

```python
# 필요한 열만 선택하고 NaN 제거
scatter_df = df[['BMI', 'Smoking']].dropna()

# 산점도 그리기
plt.figure(figsize=(8, 6))
sns.scatterplot(data=scatter_df, x='BMI', y='Smoking', color='teal', alpha=0.7)

plt.title("BMI vs Smoking 산점도")
plt.xlabel("BMI")
plt.ylabel("흡연 지수")
plt.grid(True)
plt.show()
```
